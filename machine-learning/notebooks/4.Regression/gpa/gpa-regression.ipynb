```python
# Standard includes
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
matplotlib.rc('xtick', labelsize=14) 
matplotlib.rc('ytick', labelsize=14)
```

```python
# Import data on GPAs and SAT scores
data = np.genfromtxt('gpa-data.csv', delimiter=',')
n,d = data.shape
# Three features: 0 = SAT SCORE, 1 = HS GPA, 2 = COLLEGE GPA
# We are interested in 0 and 2; let's start with a scatterplot
x = np.reshape(data[:,0], (n,1))  # predictor variable
y = data[:,2] # response variable
plt.plot(x, y, 'bo')
plt.xlabel('SAT score', fontsize=14)
plt.ylabel('College GPA', fontsize=14)
plt.show()
```

```python
# Look at feature 2 on its own; this is our response variable
plt.hist(y)
plt.xlabel('College GPA', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.show()
```

```python
np.mean(y), np.var(y)
```

```python
# Now regress y on x
regr = linear_model.LinearRegression()
regr.fit(x, y)

# Print coefficients
print('Slope, intercept: ', regr.coef_, regr.intercept_)

# Make predictions using the model and compute squared error
y_pred = regr.predict(x)
print("Mean squared error: %.2f" % mean_squared_error(y, y_pred))

# Plot outputs
plt.plot(x, y, 'bo')
plt.plot(x, y_pred, 'r-', linewidth=3)
plt.xlabel('SAT score', fontsize=14)
plt.ylabel('College GPA',fontsize=14)
plt.show()
```
