```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 1001)
```

```python
y1 = x**2
y2 = x**3
```

```python
alpha, beta = -7, 9
```

```python
np.trapz(abs( y1[((x>alpha) & (x<=beta))] - y2[((x>alpha) & (x<=beta))]), x[((x>alpha) & (x<=beta))])
```

```python
plt.plot(x, y1, 'b', linewidth = 3)
plt.plot(x, y2, 'r', linewidth = 3)
plt.gca().set_xlim([min(x), max(x)])
plt.gca().set_ylim([min(min(y1), min(y2)), max(max(y1), max(y2))])
plt.gca().fill_between(x, y1, y2, where = ((x>=alpha) & (x<=beta)), facecolor='cyan')
plt.gcf().set_size_inches([20,10])
plt.show()
```
