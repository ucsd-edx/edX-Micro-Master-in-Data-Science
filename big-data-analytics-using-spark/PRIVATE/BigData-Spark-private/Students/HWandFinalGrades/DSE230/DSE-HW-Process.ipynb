```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('DSE-Grades.csv')
```

```python
df
```

```python
hw_grades=df[:].as_matrix()[:,3:8]
```

```python
hw_grades
```

```python
min_hw=np.min(hw_grades, axis=1)
```

```python
hw_total=np.sum(hw_grades, axis=1)-min_hw
```

```python
hw_total_ser=pd.Series(hw_total)
```

```python
df['Best-4-HW']=hw_total_ser
```

```python
df
```

```python
hw_total_70=hw_total*(0.7/4)   # HW is 70% of final grade and there are best 4 out of 5 HW, each scored from 0 to 100
```

```python
hw_total_70_ser=pd.Series(hw_total_70)
df['Hw-Grade(70%)']=hw_total_70_ser
```

```python
final_grade=df[:].as_matrix()[:,8]
```

```python
final_grade*=(30./80.)   ### grade from final is in the rane 0-80 and it constitutes 30% of the final grade.
```

```python
final_grade_ser=pd.Series(final_grade)
```

```python
df['Final-Grade(30%)']=final_grade_ser
```

```python
total=hw_total_70+final_grade
```

```python
df['Total']=pd.Series(total)
```

```python
df.sort(columns='Total')
```

```python
df.to_csv('grades_with_stats.csv')
```

```python
%pylab inline
df['Total'].hist()
```

```python
Cutoffs=[(100,'A+'),
         (95,'A'),
         (88,'A-'),
         (83,'B+'),
         (75,'B')]
Cutoffs
```

```python
def letter_grade(score):
    for cut,letter in Cutoffs:
        if score>=cut:
            return letter
    return None
```

```python
Letter_Grades=[letter_grade(score) for score in list(df['Total'])]
df['Letter_Grade']=Letter_Grades
df.sort(columns='Total')
```

```python
DSE230_SectionID='870482'
df['SectionId']=DSE230_SectionID
df_columns=[u'Last Name', u'First Name', u'PID','SectionId',u'Letter_Grade']
egrade_columns=['Last Name','First Name','Student ID','SectionId','Final_Assigned_Egrade']

eGrades=pd.DataFrame()
for old,new in zip(df_columns,egrade_columns):
    eGrades[new]=df[old]
eGrades['SectionId']=DSE230_SectionID
eGrades
```

```python
eGrades.to_csv('DSE230_eGrades.csv',index=None)
```
