```python
Entrance=pd.read_csv('EntranceExam/ExamResuts.csv')
print 'Entrance Original size ',Entrance.shape
Entrance=Entrance.drop_duplicates(subset='pid',keep='last')
print 'Entrance size after dropping multiples',Entrance.shape
Entrance=Entrance.rename(columns={'pid':'PID'})
Entrance.head()
```

```python
##Clean Up PIDs
import re
PIDs=list(Entrance['PID'])
new_PIDs=[]
pattern=re.compile(r".?(\d{8})")
for pid in PIDs:
    match=pattern.match(pid)
    if not match:
        print 'MISSMATCH:',pid
    else:
        new_PIDs.append('A'+match.group(1))

Entrance['PID']=new_PIDs
Entrance.head()
```

```python
Entrance=Entrance.set_index('PID')
Combined=Combined.set_index('PID')
```

```python
Combined2=Combined.join(Entrance,lsuffix='_L',rsuffix='_R',how='left')
print Combined2.shape
Combined2.head()
```

```python
Combined2.columns
```

```python
Combined2=Combined2[[u'Student', u'Email', u'Major', u'Level', u'Date', u'Time',
       u'Wait-List', u'sum']]
```

```python
WaitingList=Combined2[Combined2['Wait-List']==True]
WaitingList.shape
```

```python
WaitingList
```

```python
#drop students that did not take the entrance Exam
WaitingList=WaitingList.dropna(axis=0,how='any')
WaitingList=WaitingList[WaitingList['sum']>4]
WaitingList.shape
```
