### Load the required libraries

```python
import pandas as pd
import numpy as np
import sklearn as sk
import urllib
from pylab import *
%matplotlib inline
```

### Read the data frame from pickle file

This file contains a dictionary with the following keys:

* UnDef = an array that gives the number of nan's in each record
* mean = the mean of all non-nan measurements
* std = the std of all non-nan measurements
* SortedVals = a sorted sample of values
* low100,high100 = bottom 1% and top 1% of the values
* low1000,high1000 = bottom .1% and top .1% of the values

```python
import pickle
Data=pickle.load(open('Weather/TMAX_Statistics.pickle','r'))
```

### Plot histogram for distribution of NaN values per record 

```python
NaNs=pd.Series(Data['UnDef'])
NaNs.hist(bins=25);
title('Distribution of NaN values per record');
xlim([0,365]);
xlabel('# NaNs');
ylabel('# Records');
```

### Plot histogram for sample values, restricted between low1000 and high1000

```python
SortedVals=pd.Series(Data['SortedVals'])
SortedVals=SortedVals[SortedVals>Data['low1000']]
SortedVals=SortedVals[SortedVals<Data['high1000']]
SortedVals.hist(bins=50)
title('Distribution of sample values');
xlim([Data['low1000'],Data['high1000']])
xlabel('Values');
ylabel('Count');
```
