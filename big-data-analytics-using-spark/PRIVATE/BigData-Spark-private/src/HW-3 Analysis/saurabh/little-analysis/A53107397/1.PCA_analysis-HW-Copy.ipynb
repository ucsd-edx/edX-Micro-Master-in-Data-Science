### Script for plotting yearly plots 

```python
def YearlyPlots(T,ttl='',yl='',xl='',y=None,x=None,size=(10,7), c=None):
    yearday=[i for i in range(1,366)]
    fig=figure(1,figsize=size,dpi=300)
    if shape(T)[0] != 365:
        raise ValueError("First dimension of T should be 365. Shape(T)="+str(shape(T)))
    if c is not None:
        plot_date(yearday,T, '-',color=c)
    else:
        plot_date(yearday,T, '-')
    # rotate and align the tick labels so they look better
    # fig.autofmt_xdate()
    ylabel(yl)
    xlabel(xl)
    if y is not None:
        ylim(y)
    if x is not None:
        xlim(x)
    grid()
    title(ttl)
```

### Plot the following 3 plots for each measurement:

1. A histogram from the sample values (from SortedVals) restricted between low100 and high100
2. Plot of mean and mean $\pm$ std
3. Number of measurements recorded each day

```python
figure(figsize=(15,30))
offset=1
for meas in STAT.keys():
    subplot(6,3,offset)
    offset+=1
    ## Histogram
    # Filter the data, keep the restricted between low100 and high100
    data_restricted = filter(lambda x: x >= STAT[meas]['low100'] \
                             and x <= STAT[meas]['high100'], STAT[meas]['SortedVals'])
    hist(data_restricted, bins=100)
    title(meas + ' restricted Hist')
    subplot(6,3,offset)
    offset+=1
    
    ## mean and mean +- std
    data_mean = STAT[meas]['Mean']
    data_std = sqrt(STAT[meas]['Var'])
    
    # Store mean, mean+std, mean-std in a matrix and plot them together
    data_concat = transpose([data_mean + data_std, data_mean - data_std])
    YearlyPlots(data_mean, c = 'r')
    YearlyPlots(data_concat, ttl = meas + ' mean +- std', c = 'b')
    subplot(6,3,offset)
    offset+=1
    
    ## Counts per day
    data_count = STAT[meas]['NE']
    YearlyPlots(data_count, ttl = meas + ' Counts', c = 'k')
```
