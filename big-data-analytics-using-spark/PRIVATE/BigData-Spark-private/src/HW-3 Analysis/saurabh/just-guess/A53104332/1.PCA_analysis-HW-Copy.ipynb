### Script for plotting yearly plots 

```python
def YearlyPlots(T,ttl='',yl='',xl='',y=None,x=None,size=(10,7), c=None):
    yearday=[i for i in range(1,366)]
    fig=figure(1,figsize=size,dpi=300)
    if shape(T)[0] != 365:
        raise ValueError("First dimension of T should be 365. Shape(T)="+str(shape(T)))
    if c is not None:
        plot_date(yearday,T, '-',color=c);
        gca().xaxis.set_major_formatter( DateFormatter('%b') )
    else:
        plot_date(yearday,T, '-', );
    # rotate and align the tick labels so they look better
    #fig.autofmt_xdate()
    ylabel(yl)
    xlabel(xl)
    if y is not None:
        ylim(y)
    if x is not None:
        xlim(x)
    grid()
    title(ttl)
```

### Plot the following 3 plots for each measurement:

1. The percentage of variance explained by top-k eigen vectors for k between 1 to 9
2. Plot of mean and mean $\pm$ std
3. Plot of top 3 eigenvectors

```python
figure(figsize=(15,30))
offset=1
for meas in STAT.keys():
    subplot(6,3,offset)
    offset+=1
    ## Your code for percentage of variance explained
    tot=sum(STAT[meas]['eigval'])
    ksum=0
    points=[0]
    for k in range(9):
        ksum+=STAT[meas]['eigval'][k]
        points.append(ksum/tot)
        
    plot(range(10),points)
    
    
    subplot(6,3,offset)
    offset+=1
    ## Your code for mean and mean +- std
    YearlyPlots(T=STAT[meas]['Mean'],c='red')
    YearlyPlots(T=STAT[meas]['Mean']+np.sqrt(STAT[meas]['Var']),c='blue')
    YearlyPlots(T=STAT[meas]['Mean']-np.sqrt(STAT[meas]['Var']),c='blue')
    title('%s mean +- std'%meas)
    
    
    subplot(6,3,offset)
    offset+=1
    ## Your code for top-3 eigenvectors
    YearlyPlots(T=asarray(STAT[meas]['eigvec'])[:,0],c='red')
    YearlyPlots(T=asarray(STAT[meas]['eigvec'])[:,1],c='green')
    YearlyPlots(T=asarray(STAT[meas]['eigvec'])[:,2],c='blue')
    title('%s Top 3 EigenVecs'%meas)
```
