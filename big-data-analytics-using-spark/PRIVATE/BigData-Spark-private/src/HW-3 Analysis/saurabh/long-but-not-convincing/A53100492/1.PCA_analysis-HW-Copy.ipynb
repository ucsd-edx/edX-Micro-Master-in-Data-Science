### Script for plotting yearly plots 

```python
def YearlyPlots(T,ttl='',yl='',xl='',y=None,x=None,size=(10,7),c=''):
    yearday=[i for i in range(1,366)]
    fig=figure(1,figsize=size,dpi=300)
    if shape(T)[0] != 365:
        raise ValueError("First dimension of T should be 365. Shape(T)="+str(shape(T)))
    plot_date(yearday,T, '-',color=c);
    plt.gca().xaxis.set_major_formatter( DateFormatter('%b') )
    # rotate and align the tick labels so they look better
#     fig.autofmt_xdate()
    ylabel(yl)
    xlabel(xl)
    if y is not None:
        ylim(y)
    if x is not None:
        xlim(x)
    grid()
    title(ttl)
```

### Plot the following 3 plots for each measurement:

1. A histogram from the sample values (from SortedVals) restricted between low1000 and high1000
2. Plot of mean and mean $\pm$ std
3. Number of measurements recorded each day

```python
figure(figsize=(15,30))
offset=1
for meas in STAT.keys():
    
    data = STAT[meas]
    
    subplot(6,3,offset)
    offset+=1
    
    hist_data = data['SortedVals']
    step = (data['high100'] - data['low100'])/100
    bins = arange(data['low100'], data['high100']+step , step)
    hist(hist_data, bins, color='blue')
    title(meas + ' restricted Hist')
    grid()
    
    subplot(6,3,offset)
    offset+=1
    
    Mean_std = vstack((data['Mean']-sqrt(data['Var']), data['Mean'], data['Mean']+sqrt(data['Var']))).transpose()
    YearlyPlots(Mean_std[:, 0],ttl=meas + ' mean +- std',yl='',xl='',y=None,x=None,size=(10,7),c='blue')
    YearlyPlots(Mean_std[:, 1],ttl=meas + ' mean +- std',yl='',xl='',y=None,x=None,size=(10,7),c='red')
    YearlyPlots(Mean_std[:, 2],ttl=meas + ' mean +- std',yl='',xl='',y=None,x=None,size=(10,7),c='blue')
    
    subplot(6,3,offset)
    offset+=1
    
    cnt = data['NE']
    YearlyPlots(cnt,ttl=meas + ' Counts',yl='',xl='',y=None,x=None,size=(10,7),c='black')
    
```
