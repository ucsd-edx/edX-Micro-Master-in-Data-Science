```python
figure(figsize=(10,7))
x=arange(0,1,0.01)
def err(p):
    return min(p,1-p)
yerr=[err(p) for p in x]
plot(x,yerr,label='err');
def H(p):
    return (-p*log2(p)-(1-p)*log2(1-p))/2. 
yH = [H(p) for p in x]
plot(x,yH,label='entropy');
yCirc = [sqrt(0.25-(.5-p)**2) for p in x]
plot(x,yCirc,label='circle');
yGini = [p*(1-p)*2 for p in x]
plot(x,yGini,label='Gini')
legend()
```

```python
figure(figsize=(10,7))
def vert(p,y,color='b'):
    plot([p,p],[0,y],color)
    plot(p,y,'o'+color)

plot(x,yerr,'b');
p=0.8; vert(p,err(p),'b')
p=0.7; vert(p,err(p),'r')
p=0.9; vert(p,err(p),'r')

plot(x,yH,'g');
p=0.8; vert(p,H(p),'b')
p=0.7; vert(p,H(p),'r')
p=0.9; vert(p,H(p),'r')
```
