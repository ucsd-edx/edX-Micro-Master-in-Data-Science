```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
# Set up Gaussian
mu = [1,1]
sigma = [[4,0],[0,1]]
rv = multivariate_normal(mu,sigma)
# Plot some samples
samples = np.random.multivariate_normal(mu,sigma,size=100)
plt.plot(samples[:,0], samples[:,1],'bo')
plt.xlim(-6,8)
plt.ylim(-6,8)
inc = 2
plt.xticks(np.arange(-6,8+inc,inc),fontsize=18)
plt.yticks(np.arange(-6,8+inc,inc),fontsize=18)
# Plot contour line
res = 200
x = np.linspace(-6,8,res)
y = np.linspace(-6,8,res)
z = np.zeros((res,res))
for i in range(0,200):
    for j in range(0,200):
        z[j,i] = rv.pdf([x[i], y[j]])
plt.contour(x,y,z,levels=[0.01],colors='r',linewidths=2.0)
# Finally, display
plt.show()
```
