```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
# Set up Gaussian
mu = [3,3]
sigma = [[1,0.85],[0.85,1]]
rv = multivariate_normal(mu,sigma)
# Plot some samples
samples = np.random.multivariate_normal(mu,sigma,size=100)
plt.plot(samples[:,0], samples[:,1],'bo')
plt.xlim(0,6)
plt.ylim(0,6)
inc = 1
plt.xticks(np.arange(0,6+inc,inc),fontsize=18)
plt.yticks(np.arange(0,6+inc,inc),fontsize=18)
# Add a line
plt.plot([0,6], [0,6], 'r-', lw=2)
# Finally, display
plt.show()
```

```python
plt.plot(samples[:,0], samples[:,1],'bo')
plt.xlim(0,6)
plt.ylim(0,6)
inc = 1
plt.xticks(np.arange(0,6+inc,inc),fontsize=18)
plt.yticks(np.arange(0,6+inc,inc),fontsize=18)
# Finally, display
plt.show()
```

```python
# Now project all points onto the line
z = (samples[:,0] + samples[:,1])/2.0
plt.plot(z,z,'bo')
plt.xlim(0,6)
plt.ylim(0,6)
inc = 1
plt.xticks(np.arange(0,6+inc,inc),fontsize=18)
plt.yticks(np.arange(0,6+inc,inc),fontsize=18)
# Add a line
plt.plot([0,6], [0,6], 'r-', lw=2)
# Finally, display
plt.show()
```

```python
# Now plot the projections in 1-d
plt.plot(z * np.sqrt(2),np.zeros(100),'bo')
plt.xlim(0,9)
plt.ylim(-1,1)
inc = 1
plt.xticks(np.arange(0,9+inc,inc),fontsize=18)
plt.yticks([])
# Add a line
plt.plot([0,9], [0,0], 'r-', lw=2)
# Finally, display
#plt.axis('off')
plt.show()
```
